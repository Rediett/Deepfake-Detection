```python
import os
from tensorflow.keras.models import load_model
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
label = sorted(["fake","real"])
new_model = load_model('./models/Deepfake_Detection.keras')
```

```python
test_path = '/Users/rediettadesse/Desktop/AI_projects/Deepfake Detection/test_imgs/resized-real-160.png.png'
image = tf.keras.utils.load_img(test_path,color_mode="grayscale",target_size=(100,100))
input_arr = tf.keras.utils.img_to_array(image)
input_arr = np.array([input_arr])/255.0  # Convert single image to a batch.

predictions = new_model.predict(input_arr)
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.title("Predicted : "+label[round(predictions[0][0])])
plt.show()
```
